- ### Identify underperforming brands that require promotional or pricing adjustments.

```python
import time
conn = sqlite3.connect("Inventory.db")
start = time.time()
VendorSalesSummary = pd.read_sql_query("""WITH sales_summary AS(
                  SELECT brand,
                  vendornumber,
                  description,
                  SUM(salesquantity) AS TotalSalesQuantity,
                  SUM(salesdollars) AS TotalSalesDollars,
                  SUM(excise) AS TotalExcise
                  FROM sales
                  GROUP BY brand,vendornumber
                  ),
                  Purchase_summary AS (
                  SELECT p.brand,
                  p.Vendornumber,
                  AVG(pp.purchaseprice) AS TotalPurchasePrice,
                  SUM(p.quantity) AS TotalPurchaseQuantity,
                  SUM(p.dollars) AS TotalPurchaseDollars
                  FROM  purchases as p 
                  LEFT JOIN  purchase_prices as pp
                  ON p.brand=pp.brand and p.vendornumber=pp.vendornumber
                  GROUP BY p.brand,p.vendornumber
                  ),
                  vendor_summary AS (
                  SELECT vendornumber,
                  SUM(freight) AS TotalFreightCost
                  FROM vendor_invoice
                  GROUP BY vendornumber
                  )
                  SELECT s.brand,
                  s.vendornumber,
                  s.description,
                  s.TotalSalesQuantity,
                  s.TotalSalesDollars,
                  s.TotalExcise,
                  p.TotalPurchasePrice,
                  p.TotalPurchaseQuantity,
                  p.TotalPurchaseDollars,
                  vi.TotalFreightCost
                  FROM 
                  sales_summary as s
                  LEFT JOIN  Purchase_summary as p ON s.brand=p.brand and s.vendornumber=p.vendornumber
                  LEFT JOIN vendor_summary AS vi ON s.vendornumber=vi.vendornumber
                  """,conn)
end= time.time()
total_time = round((end - start) / 60,2)
print(f'Query took {total_time} seconds to complete')
```

```python
conn = sqlite3.connect("Vendor_sales.db")
cursor=conn.cursor()
cursor.execute("DROP TABLE IF EXISTS VendorSales;")
cursor.execute("""CREATE TABLE VendorSales(
                  brand INT,
                  vendornumber INT,
                  description VARCHAR(50),
                  TotalSalesQuantity INT,
                  TotalSalesDollars DECIMAL(12,6),
                  TotalExcise DECIMAL(12,6),
                  TotalPurchasePrice DECIMAL(12,6),
                  TotalPurchaseQuantity INT,
                  TotalPurchaseDollars DECIMAL(12,6),
                  TotalFreightCost DECIMAL(12,6),
                  PRIMARY KEY (brand,vendornumber)
                  );""")
```

```python
try:
    VendorSalesSummary.to_sql("VendorSales",conn,if_exists='replace',index=False)
except Exception as e:
    logging.error(f"error inserting values {e}")
else:
    logging.info('--records inserted successfully-- ')
```
